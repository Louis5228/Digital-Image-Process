```
import os
from google.colab import drive
from matplotlib import pyplot as plt
import numpy as np
import cv2
import math
drive.mount('/content/gdrive')

os.chdir("/content/gdrive/My Drive/影像處理") #更改路徑
```

```
img = cv2.imread('Bird feeding 3 low contrast.tif', cv2.IMREAD_GRAYSCALE)
show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
plt.title('Input Figure')
plt.show()
```

```
plt.hist(img.ravel(),256,[0,256])
plt.title('Original histogram')
plt.show()
```

```
new_img = ((np.arctan((show_img - 128.0) / 32.0) + 1.3258) * 96.234).astype(int)
plt.imshow(new_img)
plt.title('Output Figure')
plt.show()
```

```
plt.hist(new_img.ravel(),256,[0,256])
plt.title('Output histogram')
plt.show()
```

```
x = np.arange(0, 255, 0.1)
y = (np.arctan((x - 128.0) / 32.0) + 1.3258) * 96.234
plt.plot(x, y, label="s = T(r)")
plt.xlabel("r")
plt.ylabel("s")
plt.title('Figure of s = T(r)')
plt.legend()
plt.show()
```
